```python
X = df.loc[:, features]
y = df.loc[:, ['Adj_Close']]
```

Split Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.15,random_state=42)
```

Standardize data

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

Part 2 PCA on Treasury Yield dataset -
Compute and display the explained variance ratio for all components, then recalculate and display on 
n_components=3.
What is the cumulative explained variance of the 3 component version.

```python
cov_mat = np.cov(X_train_std.T)
eigen_vals, eigen_vecs = np.linalg.eig(cov_mat)

print('\nEigenvalues \n%s' % eigen_vals)
```

```python
tot = sum(eigen_vals)
var_exp = [(i / tot) for i in sorted(eigen_vals, reverse=True)]
cum_var_exp = np.cumsum(var_exp)
```

```python
cum_var_exp
```

Total and explained variance graph for all components

```python
import matplotlib.pyplot as plt


plt.bar(range(1, 31), var_exp, alpha=0.5, align='center',
        label='individual explained variance')
plt.step(range(1, 31), cum_var_exp, where='mid',
         label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
# plt.savefig('images/05_02.png', dpi=300)
plt.show()
```

PCA using scikit-learn for 3 components

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train_std)
X_test_pca = pca.transform(X_test_std)

pca.explained_variance_ratio_
```

```python
plt.bar(range(1, 4), pca.explained_variance_ratio_, alpha=0.5, align='center')
plt.step(range(1, 4), np.cumsum(pca.explained_variance_ratio_), where='mid')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal components')

plt.show()
```

Part 3: Linear regression v. SVM regressor - baseline
using all features below

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
import time
 
# record start time
start = time.time()
 
a = 0
for i in range(1000):
    a += (i**100)


# Fit linear regression model to dataset with 30 features
reg_30 = LinearRegression()
reg_30.fit(X_train, y_train)

# Predict on train set and calculate R2 and RMSE
y_pred_train_30 = reg_30.predict(X_train)
r2_train_30 = r2_score(y_train, y_pred_train_30)
rmse_train_30 = np.sqrt(mean_squared_error(y_train, y_pred_train_30))
print(rmse_train_30,r2_train_30)
# Predict on test set and calculate R2 and RMSE
y_pred_test_30 = reg_30.predict(X_test)
r2_test_30 = r2_score(y_test, y_pred_test_30)
rmse_test_30 = np.sqrt(mean_squared_error(y_test, y_pred_test_30))
print(rmse_test_30,r2_test_30)

# record end time
end = time.time()
 
linear_30_comp_time= (end-start) * 10**3

# print the difference 

print("The time of execution of above program is :",
      (end-start) * 10**3, "ms")
```

Linear regression on PCA dataset with 3 features

```python
pca = PCA(n_components=3)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
reg_pca = LinearRegression()
reg_pca.fit(X_train_pca, y_train)

# Predict on train set and calculate R2 and RMSE
y_pred_train_pca = reg_pca.predict(X_train_pca)
r2_train_pca = r2_score(y_train, y_pred_train_pca)
rmse_train_pca = np.sqrt(mean_squared_error(y_train, y_pred_train_pca))
print(rmse_train_pca,r2_train_pca)
# Predict on test set and calculate R2 and RMSE
y_pred_test_pca = reg_pca.predict(X_test_pca)
r2_test_pca = r2_score(y_test, y_pred_test_pca)
rmse_test_pca = np.sqrt(mean_squared_error(y_test, y_pred_test_pca))
print(rmse_test_pca,r2_test_pca)
```

SVM regressor on both datasets
