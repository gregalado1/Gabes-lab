```python
features = ['Sales/Revenues', 'Gross Margin', 'EBITDA', 'EBITDA Margin',
       'Net Income Before Extras', 'Total Debt', 'Net Debt', 'LT Debt',
       'ST Debt', 'Cash', 'Free Cash Flow', 'Total Debt/EBITDA',
       'Net Debt/EBITDA', 'Total MV', 'Total Debt/MV', 'Net Debt/MV',
       'CFO/Debt', 'CFO', 'Interest Coverage', 'Total Liquidity',
       'Current Liquidity', 'Current Liabilities', 'EPS Before Extras', 'PE',
       'ROA', 'ROE']
```

```python
X = df.loc[:, features]
y = df.loc[:, ['InvGrd']].values.ravel()
```

```python
from sklearn.preprocessing import LabelEncoder
# categorical variables
le = LabelEncoder()
y = le.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = \
train_test_split(X, y, test_size=0.15,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
from sklearn.linear_model import ElasticNetCV
from sklearn.feature_selection import SelectFromModel
```

```python
# Train Elastic Net model with cross-validation
elastic_net = ElasticNetCV(l1_ratio=[.1, 1], cv=10)
elastic_net.fit(X_train_std, y_train)

#most important features
selector = SelectFromModel(elastic_net, prefit=True)
X_train_selected = selector.transform(X_train_std)
X_test_selected = selector.transform(X_test_std)

selected_features = X.columns[selector.get_support()]
print("Selected features:", selected_features)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=1)
X_train_lda = lda.fit_transform(X_train_selected, y_train)
X_test_lda = lda.transform(X_test_selected)

print('Explained variance ratio:', lda.explained_variance_ratio_)
```

```python
#3) Model fitting and evaluation

import pandas as pd
from sklearn.pipeline import Pipeline
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score


#pipeline LDA and logistic regression
pipe_lr_lda = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)),
                        ('clf', LogisticRegression(random_state=42))])

# pipe LDA and support vector machine
pipe_svc_lda = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)),
                         ('clf', SVC(random_state=42))])

# pipe LDA and random forest
pipe_rf_lda = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)),
                        ('clf', RandomForestClassifier(random_state=42))])

pipe_lr_lda.fit(X_train, y_train)
pipe_svc_lda.fit(X_train, y_train)
pipe_rf_lda.fit(X_train, y_train)

#scores
y_pred_lr_lda = pipe_lr_lda.predict(X_test)
y_pred_svc_lda = pipe_svc_lda.predict(X_test)
y_pred_rf_lda = pipe_rf_lda.predict(X_test)

acc_lr_lda = accuracy_score(y_test, y_pred_lr_lda)
acc_svc_lda = accuracy_score(y_test, y_pred_svc_lda)
acc_rf_lda = accuracy_score(y_test, y_pred_rf_lda)

print(f'Accuracy LR-LDA: {acc_lr_lda:.4f}')
print(f'Accuracy SVM-LDA: {acc_svc_lda:.4f}')
print(f'Accuracy RF-LDA: {acc_rf_lda:.4f}')
```

```python
#4
from sklearn.model_selection import GridSearchCV

# Define the hyperparameter grid for each pipeline
param_grid_lr_lda = {'lda__n_components': [1, 2, 3],
                     'clf__C': [0.1, 1, 10]}

param_grid_svc_lda = {'lda__n_components': [1, 2, 3],
                      'clf__C': [0.1, 1, 10],
                      'clf__kernel': ['linear', 'rbf']}

param_grid_rf_lda = {'lda__n_components': [1, 2, 3],
                     'clf__n_estimators': [50, 100, 200],
                     'clf__max_depth': [3, 5, None]}

pipelines = {'LR-LDA': (pipe_lr_lda, param_grid_lr_lda),
             'SVM-LDA': (pipe_svc_lda, param_grid_svc_lda),
             'RF-LDA': (pipe_rf_lda, param_grid_rf_lda)}

# Fit pipelines grid search and cross-validation
results = {}
for pipeline_name, (pipeline, param_grid) in pipelines.items():
    grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    results[pipeline_name] = {'best_params': grid_search.best_params_,
                              'cv_results': grid_search.cv_results_,
                              'test_score': grid_search.score(X_test, y_test)}

for pipeline_name, result in results.items():
    print(pipeline_name)
    print('Best parameters:', result['best_params'])
    print('Test accuracy:', result['test_score'])
```

```python
#5 ensembling
from sklearn.ensemble import AdaBoostClassifier, BaggingClassifier
```

```python
# pipeline
pipe_lr_lda_ada = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)),   ('clf', AdaBoostClassifier(base_estimator=LogisticRegression(random_state=42), random_state=42))])

param_grid_lr_lda_ada = {'lda__n_components': [1],
                         'clf__base_estimator__C': [0.1, 1, 10],
                         'clf__n_estimators': [50, 100, 200]}

# pipeLDA, SVM and AdaBoostClassifier
pipe_svc_lda_ada = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)), ('clf', AdaBoostClassifier(base_estimator=SVC(random_state=42), algorithm='SAMME'))])

param_grid_svc_lda_ada = {'lda__n_components': [1],
                          'clf__base_estimator__C': [0.1, 1, 10],
                          'clf__base_estimator__kernel': ['linear', 'rbf'],
                          'clf__n_estimators': [50, 100, 200]}

# pipe LDA, Random Forest and BaggingClassifier
pipe_rf_lda_bag = Pipeline([('lda', LinearDiscriminantAnalysis(n_components=1)),('clf', BaggingClassifier(RandomForestClassifier(random_state=42)))])

param_grid_rf_lda_bag = {'lda__n_components': [1],
                         'clf__base_estimator__n_estimators': [50, 100, 200],
                         'clf__base_estimator__max_depth': [3, 5, None],
                         'clf__n_estimators': [50, 100, 200]}

pipelines_ensemble = {'LR-LDA-AdaBoost': (pipe_lr_lda_ada, param_grid_lr_lda_ada),
                      'SVM-LDA-AdaBoost': (pipe_svc_lda_ada, param_grid_svc_lda_ada),
                      'RF-LDA-Bagging': (pipe_rf_lda_bag, param_grid_rf_lda_bag)}

results_ensemble = {}
for pipeline_name, (pipeline, param_grid) in pipelines_ensemble.items():
    grid_search = GridSearchCV(pipeline, param_grid, cv=10, scoring='accuracy')
    grid_search.fit(X_train, y_train)
    results_ensemble[pipeline_name] = {'best_params': grid_search.best_params_,
                                       'cv_results': grid_search.cv_results_,
                                       'test_score': grid_search.score(X_test, y_test)}

for pipeline_name, result in results_ensemble.items():
    print(pipeline_name)
    print('Best parameters:', result['best_params'])
    print('Test accuracy:', result['test_score'])
```
